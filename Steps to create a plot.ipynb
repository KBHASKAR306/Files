```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
ramdomNumber = np.random.rand(10)
print(ramdomNumber)
```

```python
style.use('ggplot')
plt.plot(ramdomNumber,'g',label = 'line one', linewidth = 2)
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('First Plot')
plt.legend()
plt.show()
```
